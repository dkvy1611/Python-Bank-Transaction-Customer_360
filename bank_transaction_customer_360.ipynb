```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df['Ngày giao dịch'] = pd.to_datetime(df['Ngày giao dịch'], format='%d/%m/%Y %H:%M')

# Tạo cột 'Tuần' để nhóm theo tuần
df['Tuần'] = df['Ngày giao dịch'].dt.to_period('W').apply(lambda r: r.start_time)

# Tính tần suất giao dịch và số tiền trung bình hàng tuần
weekly_stats = df.groupby('Tuần').agg(
    Tần_suất=('Mã giao dịch', 'count'),  # Đếm số lượng giao dịch
    Số_tiền_trung_bình=('Số tiền', 'mean')  # Tính trung bình số tiền
).reset_index()

# Vẽ biểu đồ
fig, ax1 = plt.subplots(figsize=(12, 6))

# Biểu đồ đường: số tiền trung bình
ax1.plot(weekly_stats['Tuần'], weekly_stats['Số_tiền_trung_bình'], color='blue', marker='o', label='Số tiền trung bình (triệu đồng)')
ax1.set_ylabel('Số tiền trung bình (triệu đồng)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Biểu đồ cột: tần suất giao dịch
ax2 = ax1.twinx()
ax2.bar(weekly_stats['Tuần'], weekly_stats['Tần_suất'], color='orange', alpha=0.6, label='Tần suất giao dịch')
ax2.set_ylabel('Tần suất giao dịch', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

ax1.set_xlabel('Tuần')
plt.xticks(rotation=45)

fig.suptitle('Tần suất và số tiền giao dịch trung bình hàng tuần', fontsize=14)
fig.tight_layout()
plt.legend(loc='upper left')

plt.show()
```

**III. Hồ sơ tài chính**: 
- Kích thước trung bình của khoản gửi/rút tiền.
- Tổng giá trị giao dịch theo thời gian.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df['Ngày giao dịch'] = pd.to_datetime(df['Ngày giao dịch'], format='%d/%m/%Y %H:%M')

# Tạo cột 'Tuần' để nhóm theo tuần
df['Tuần'] = df['Ngày giao dịch'].dt.to_period('W').apply(lambda r: r.start_time)

# Tính tổng giá trị giao dịch hàng tuần
weekly_stats = df.groupby('Tuần').agg(
    Tổng_giá_trị_giao_dịch=('Số tiền', 'sum')  # Tổng số tiền
).reset_index()

# Tính giá trị trung bình
average_value = weekly_stats['Tổng_giá_trị_giao_dịch'].mean()

# Vẽ biểu đồ
plt.figure(figsize=(12, 6))
sns.barplot(x='Tuần', y='Tổng_giá_trị_giao_dịch', data=weekly_stats, color='skyblue', label='Tổng giá trị giao dịch')

# Thêm đường trung bình
plt.axhline(y=average_value, color='red', linestyle='--', label=f'Giá trị trung bình ({average_value:,.0f})')

plt.title('Tổng giá trị giao dịch theo tuần', fontsize=14)
plt.xlabel('Tuần', fontsize=12)
plt.ylabel('Tổng giá trị giao dịch', fontsize=12)
plt.xticks(rotation=45)

# Thêm giá trị trên từng cột
for p in plt.gca().patches:
    plt.gca().annotate(
        f'{p.get_height():,.0f}',  # Hiển thị số tổng giao dịch
        (p.get_x() + p.get_width() / 2., p.get_height()),  # Vị trí
        ha='center', va='baseline', fontsize=10, color='black', xytext=(0, 5), textcoords='offset points'
    )

plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

Cần có thêm thông tin giao dịch của các khách hàng khác để có thể nhận diện khách hàng có giá trị cao và xếp hạng khách hàng theo giá trị tiền tệ.

**V. Giá trị tài chính của từng khách hàng**

```python
import pandas as pd
import re

# Hàm trích xuất tên người chuyển tiền
def extract_name(transaction_detail):
    match = re.search(r'(QR - )?([A-Za-z\s]+)(?=\s+chuyen tien)', transaction_detail)
    if match:
        return match.group(2).strip()  # Trả về tên
    return None

# Áp dụng hàm để tạo cột 'Tên người chuyển tiền'
df['Tên người chuyển tiền'] = df['Nội dung chuyển khoản'].apply(extract_name)

# Lọc bỏ giao dịch có "PHUNG THI HANG" (người nhận)
df_filtered = df[df['Tên người chuyển tiền'] != 'PHUNG THI HANG']

# Tính tổng số tiền giao dịch và giá trị giao dịch trung bình cho mỗi người
summary = df_filtered.groupby('Tên người chuyển tiền').agg(
    Số_giao_dịch=('Số tiền', 'count'),
    Tổng_số_tiền=('Số tiền', 'sum'),
    Giá_trị_giao_dịch_trung_bình=('Số tiền', 'mean')
).reset_index()

# Tính giá trị trung bình
avg_transactions = summary['Số_giao_dịch'].mean()
avg_total_value = summary['Tổng_số_tiền'].mean()

# Phân nhóm khách hàng
def assign_group(row):
    if row['Số_giao_dịch'] < avg_transactions and row['Tổng_số_tiền'] > avg_total_value:
        return 'Nhóm 1: Ít lần - Tổng cao'
    elif row['Số_giao_dịch'] >= avg_transactions and row['Tổng_số_tiền'] <= avg_total_value:
        return 'Nhóm 2: Nhiều lần - Tổng thấp'
    elif row['Số_giao_dịch'] < avg_transactions and row['Tổng_số_tiền'] <= avg_total_value:
        return 'Nhóm 3: Ít lần - Tổng thấp'
    elif row['Số_giao_dịch'] >= avg_transactions and row['Tổng_số_tiền'] > avg_total_value:
        return 'Nhóm 4: Nhiều lần - Tổng cao'

summary['Nhóm'] = summary.apply(assign_group, axis=1)

# Sắp xếp theo nhóm và tổng số tiền giao dịch desc
summary_sorted = summary.sort_values(by=['Nhóm', 'Tổng_số_tiền'], ascending=[True, False])

# Hiển thị bảng
from IPython.display import display, HTML
display(HTML(summary_sorted.to_html(classes='table table-striped table-bordered', escape=False)))
```

  - Đây là nhóm khách hàng quan trọng cần giữ chân bằng mọi cách, vì họ đóng góp lớn vào doanh thu.
  - Cần cung cấp các chương trình chăm sóc đặc biệt, như giảm phí, tăng lợi ích cho giao dịch thường xuyên.
  - Khuyến khích thêm các sản phẩm dịch vụ bổ sung để duy trì và phát triển quan hệ dài hạn.

**HÀNH ĐỘNG ĐỀ XUẤT**:
1. **Phân tích chi tiết** từng nhóm để hiểu rõ hơn về đặc điểm và nhu cầu.
2. **Định hướng chiến lược** cho từng nhóm:
   - **Nhóm giá trị cao**: Tập trung chăm sóc và giữ chân.
   - **Nhóm giá trị thấp nhưng tiềm năng**: Đầu tư khuyến khích tăng trưởng.
   - **Nhóm không tiềm năng**: Cân nhắc tối ưu hóa chi phí tiếp cận.
3. **Theo dõi thường xuyên** qua các chỉ số như RFM (Recency, Frequency, Monetary) để xác định thay đổi trong hành vi của khách hàng.
